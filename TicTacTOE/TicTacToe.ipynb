### 6. take position as input

```python
from IPython.display import clear_output
```

```python
def take_position():
    """ taking postion of the marker as input """
    # creating a placeholder for the position
    display(li)
    position = None
    # a loop till currect input is in place
    while not position:
        # taking input - position
        position = input()
        # checking if the input is int
        if not position.isnumeric():
            # reseting the position if input not int
            position = None
        # checking input if it is in range 1 to 9
        elif not int(position) in range(1, 10):
            # reseting the position if input is not in range
            position = None
        # checking for blank space in list
        elif not li[int(position)] ==" ":
            position = None
        clear_output()
        
    # returning the int type position as input is always in string
    return int(position)
```

```python
def checkDraw():
    """ check for empty space from range 1-9"""
    return not " " in li[1:]
```

```python
def ChangeMarker(marker):
    """exchanging marker O - X """
    if marker == "X":
        return "O"
    elif marker == "O":
        return "X"
```
